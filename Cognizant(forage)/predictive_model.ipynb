```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sales_df = pd.read_csv('./inputs/sales.csv')
sales_df.drop(columns=["Unnamed: 0"], inplace=True, errors='ignore')

sales_df
```

```python
stock_df = pd.read_csv('./inputs/sensor_storage_data.csv')
stock_df.drop(columns=["Unnamed: 0"], inplace=True, errors='ignore')

stock_df
```

```python
temp_df = pd.read_csv('./inputs/sensor_temp.csv')
temp_df.drop(columns=["Unnamed: 0"], inplace=True, errors='ignore')

temp_df
```

```python
def format_timestamp(df):
    df['formatted_timestamp'] = pd.to_datetime(df['timestamp'],format='%Y-%m-%d %H:%M:%S')
    df.drop('timestamp',inplace=True,axis=1)
    return df
```

```python
sales_df = format_timestamp(sales_df)
stock_df =format_timestamp(stock_df)
temp_df = format_timestamp(temp_df)
```

```python
sales_df['formatted_timestamp']
```

```python
def set_time_to_hr(df):
    df['formatted_timestamp'] = df['formatted_timestamp']
```

```python
stock_df
```

```python
from datetime import timedelta
def round_to_nearest_hour(dt):
    return (dt.replace(second=0, microsecond=0, minute=0, hour=dt.hour)
            + timedelta(hours=dt.minute//30))

sales_df['formatted_timestamp'] = sales_df['formatted_timestamp'].map(
    lambda x: round_to_nearest_hour(x).strftime("%Y-%m-%d %H:00:00")
)

stock_df['formatted_timestamp'] = stock_df['formatted_timestamp'].map(
    lambda x: round_to_nearest_hour(x).strftime("%Y-%m-%d %H:00:00")
)

temp_df['formatted_timestamp'] = temp_df['formatted_timestamp'].map(
    lambda x: round_to_nearest_hour(x).strftime("%Y-%m-%d %H:00:00")
)
```

```python
sales_df
```

```python
sales_agg = sales_df.groupby(['formatted_timestamp','product_id']).agg({'quantity':'sum'}).reset_index()
sales_agg
```

```python
stock_agg = stock_df.groupby(
    ['formatted_timestamp','product_id']
    ).agg({'estimated_stock_pct':'mean'}).reset_index()
stock_agg
```

```python
temp_agg = temp_df.groupby(
    ['formatted_timestamp']
).agg({'temperature':'mean'}).reset_index()
temp_agg
```

```python
merged_df = sales_agg.merge(stock_agg,how='right',on=['formatted_timestamp','product_id'])
```

```python
merged_df = merged_df.merge(temp_agg,how='left',on='formatted_timestamp')
merged_df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,GradientBoostingRegressor
```

```python
y = merged_df['estimated_stock_pct']
X = merged_df.drop('estimated_stock_pct',axis=1)


scaler = StandardScaler()

X_processed = scaler.fit_transform(X)

X_train,X_test,y_train,y_test = train_test_split(X_processed,y,random_state=123,shuffle=True)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
result = {}

def model_evalueation(model,train = X_train,test = X_test):
    model = model
    model.fit(train,y_train)
    preds = model.predict(test)
    score = mean_absolute_error(y_test,preds)
    return score

models = [KNeighborsRegressor(),
          SVR(kernel='rbf'),
          RandomForestRegressor(),
          AdaBoostRegressor(),
          GradientBoostingRegressor()]

for model in models:
    model_score = model_evalueation(model)
    model_name = model.__str__()
    result[model_name] = model_score
```

```python
keys = list(result.keys())
vals = [result[k] for k in keys]
plt.show()
sns.barplot(x=keys, y=vals, palette="viridis")
plt.xticks(rotation=45)
for i, val in enumerate(vals):
    plt.text(i, val+.001, str(round(val,3)), ha='center')
    
plt.xlabel('---- Models ---->')
plt.ylabel('----- Mean-Squared-Error ----->')
plt.show()
    
```

```python
model = GradientBoostingRegressor()
model.fit(X_train,y_train)
preds = model.predict(X_test)
score = mean_squared_error(y_test,preds)
print(score)
```

```python
feature_importances = model.feature_importances_

# Define feature names (assuming we have 28 features)
feature_names = X.columns.tolist()

# Create a dictionary of feature importances
importance_dict = dict(zip(feature_names, feature_importances))

# Sort the dictionary by importance
sorted_importance = dict(sorted(importance_dict.items(), key=lambda item: item[1], reverse=True))

# Prepare data for pie chart
labels = list(sorted_importance.keys())
sizes = list(sorted_importance.values())

# Plot the pie chart
plt.figure(figsize=(12, 8))
wedges, texts, autotexts = plt.pie(
    sizes, labels=labels, autopct='%1.1f%%', startangle=140, pctdistance=0.85, colors=sns.color_palette('husl', len(labels))
)

# Beautify the pie chart
for text in texts:
    text.set_fontsize(10)
    text.set_rotation(45)

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(10)

plt.title('Relative Importance of Features\nwhen Predicting Target Variable', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
